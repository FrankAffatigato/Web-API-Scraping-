```python
from bs4 import BeautifulSoup as bs
import requests as req
url = "https://www.programmableweb.com/apis/directory"
```

```python
r = req.get(url)
c = r.content

soup = bs(c, "html.parser")
all = soup.find_all("td", {"class":"views-field views-field-pw-version-title"})
#print(soup.prettify)
```

```python
l = []
for item in all:
  d = {}
  d["API Name"]=item.text.replace(' ', '')
  d["Description"]=item.find("td", {"class","views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8"})
  print(d)
```

```python
print(soup.find("td", {"class":"views-field views-field-pw-version-title"}).text)
```

```python
l = []
first = soup.find_all("tr", {"class":"odd views-row-first"})
all = soup.find_all("tr")
for item in all:
  d = {}
  try:
    d["API Name"]=item.find("td", {"class":"views-field views-field-pw-version-title"}).text.replace(" ", "")
  except:
    None
  try:
    d["Description"]=item.find("td", {"class","views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8"}).text
  except:
    None
  try:
    d["Category"]=item.find("td", {"class":"views-field views-field-field-article-primary-category"}).text.replace(" ", "")
  except:
    None
  try:
    d["Followers"]=item.find("td", {"class":"views-field views-field-flag-follow-api-count"}).text.replace(" ", "")
  except:
    None
  try:
    d["Versions"]=item.find("td", {"class":"views-field views-field-pw-version-links"}).text.replace(" ", "")
  except:
    None
  print(d)
  l.append(d)
```

```python
import pandas as pd
df = pd.DataFrame(l)
df = df.iloc[1:]
display(df)
```

```python
df.to_csv("API_Web_Scraping.csv")
```
